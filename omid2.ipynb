```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Parameters
save_path = "/Users/omid/Downloads/E5LG_CRBL_crux1_20250516_EBC-325/saved_trials.h5"
pre_time = 100   # before LED_on
post_time = 600  # after LED_on
aligned_window = (-pre_time, post_time)
pdf_output_path = "/Users/omid/Downloads/E5LG_CRBL_crux1_20250516_EBC-325/aligned_trials.pdf"

# Load trial list
with h5py.File(save_path, "r") as f:
    trial_ids = list(f["trial_id"].keys())
    trial_ids = sorted(trial_ids, key=lambda x: int(x))  # ensure numerical order

n_trials = len(trial_ids)
n_cols = 5
n_rows = int(np.ceil(n_trials / n_cols))

# Create figure
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 2.5), sharex=True, sharey=True)
axes = axes.flatten()

probe_ids   = []
normal_ids  = []

for idx, trial_id in enumerate(trial_ids):
    ax = axes[idx]
    with h5py.File(save_path, "r") as f:
        grp = f[f"trial_id/{trial_id}"]
        time = grp["time"][:]
        dff = grp["dff"][:]
        led_on = np.array(grp["LED_on"][:]).squeeze()
        led_off = np.array(grp["LED_off"][:]).squeeze()
        airpuff = grp["AirPuff"][:]
        is_probe = bool(grp["probe"][()]) if "probe" in grp else False

        # Align time
        aligned_time = time - led_on

        # Check for time length mismatch
        if len(aligned_time) != dff.shape[1]:
            print(f"Warning: Trial {int(trial_id) + 1} has time ({len(aligned_time)}) ≠ dff ({dff.shape[1]}). Trimming to match.")
            min_len = min(len(aligned_time), dff.shape[1])
            aligned_time = aligned_time[:min_len]
            dff = dff[:, :min_len]

        # Apply time window mask
        mask = (aligned_time >= aligned_window[0]) & (aligned_time <= aligned_window[1])
        if np.sum(mask) == 0:
            continue

        aligned_time = aligned_time[mask]
        dff = dff[:, mask]
        mean_dff = dff.mean(axis=0)
        sem_dff = dff.std(axis=0) / np.sqrt(dff.shape[0])

        # Plot trace + SEM
        ax.plot(aligned_time, mean_dff, color='black')
        ax.fill_between(aligned_time, mean_dff - sem_dff, mean_dff + sem_dff, color='gray', alpha=0.3)

        # LED window
        ax.axvspan(0, led_off - led_on, color='gray', alpha=0.2)

        # AirPuff window with color-coding
        if len(airpuff) == 2:
            puff_start_aligned = airpuff[0] - led_on
            puff_end_aligned = airpuff[1] - led_on
            delay = round(puff_start_aligned)
            
            if is_probe:                       # ← probe colouring
                puff_color = "red"
                probe_ids.append(int(trial_id))

            elif 390 <= delay <= 410:
                puff_color = 'green'
            elif 190 <= delay <= 210:
                puff_color = 'blue'
            else:
                puff_color = 'orange'

            ax.axvspan(puff_start_aligned, puff_end_aligned, color=puff_color, alpha=0.2)

        # Updated trial title (1-based indexing)
        title = f"Trial {int(trial_id)+1}"
        if is_probe:
            title += " (P)"          # mark probe
        ax.set_title(title)
        ax.set_xlim(aligned_window)
        ax.tick_params(labelbottom=True, labelleft=True)

        # Add axis labels only to bottom-left plots
        if idx // n_cols == n_rows - 1:
            ax.set_xlabel("Time from LED onset (ms)")
        if idx % n_cols == 0:
            ax.set_ylabel("dF/F")

# Remove unused axes
for i in range(len(trial_ids), len(axes)):
    fig.delaxes(axes[i])

# Final formatting
fig.suptitle("Trials aligned to LED onset", fontsize=16)
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig(pdf_output_path, dpi=300)
#plt.show()

print(f"PDF saved to: {pdf_output_path}")
```

```python
import h5py, numpy as np, matplotlib.pyplot as plt, math

# ───────── parameters & paths ───────────────────────────────────────────────
session_dir   = "/Users/omid/Downloads/E5LG_CRBL_crux1_20250516_EBC-325"
save_path     = f"{session_dir}/saved_trials.h5"
pdf_out       = f"{session_dir}/probe_and_next.pdf"

pre_time, post_time = 100, 600
win            = (-pre_time, post_time)
n_cols         = 5                         # grid width
# ─────────────────────────────────────────────────────────────────────────────

# ---------- 1. find probe and probe+1 trial-ids ------------------------------
with h5py.File(save_path, "r") as f:
    all_ids   = sorted(map(int, f["trial_id"].keys()))
    probes    = [tid for tid in all_ids
                 if "probe" in f[f'trial_id/{tid}'] and
                    bool(f[f'trial_id/{tid}/probe'][()])]
# add probe+1 if it exists
plot_ids = []
for pid in probes:
    plot_ids.append(pid)
    if pid + 1 in all_ids:
        plot_ids.append(pid + 1)
plot_ids = sorted(dict.fromkeys(plot_ids))      # remove duplicates, keep order
print(f"Plotting {len(plot_ids)} trials → {plot_ids}")

# ---------- 2. prepare figure grid ------------------------------------------
n_plots  = len(plot_ids)
n_rows   = math.ceil(n_plots / n_cols)
fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(n_cols*4, n_rows*2.5),
    sharex=True, sharey=True
)
axes = axes.flatten()

# ---------- 3. draw each selected trial -------------------------------------
for idx, tid in enumerate(plot_ids):
    ax   = axes[idx]
    with h5py.File(save_path, "r") as f:
        grp  = f[f"trial_id/{tid}"]
        time = grp["time"][:]
        dff  = grp["dff"][:]
        led_on  = grp["LED_on"][0]
        led_off = grp["LED_off"][0]
        puff    = grp["AirPuff"][:]
        probe   = bool(grp["probe"][()]) if "probe" in grp else False

    # align & trim
    t_al = time - led_on
    if t_al.size != dff.shape[1]:
        m = min(t_al.size, dff.shape[1])
        t_al, dff = t_al[:m], dff[:, :m]

    mask = (t_al >= win[0]) & (t_al <= win[1])
    t_al = t_al[mask]
    dff  = dff[:, mask]

    mean = dff.mean(axis=0)
    sem  = dff.std(axis=0) / np.sqrt(dff.shape[0])

    ax.plot(t_al, mean, color="black")
    ax.fill_between(t_al, mean-sem, mean+sem, color="gray", alpha=.3)

    # LED window
    ax.axvspan(0, led_off-led_on, color="gray", alpha=.2)

    # Air-puff window (real or inferred)
    if len(puff) == 2:
        puff_on  = puff[0] - led_on
        puff_off = puff[1] - led_on
        delay    = round(puff_on)
        if probe:
            colour = "red"
        elif 190 <= delay <= 210:
            colour = "blue"
        elif 390 <= delay <= 410:
            colour = "green"
        else:
            colour = "orange"
        ax.axvspan(puff_on, puff_off, color=colour, alpha=.25)

    # titles & labels
    title = f"Trial {tid+1}"
    if probe: title += " (P)"
    ax.set_title(title)
    ax.set_xlim(win)
    if idx // n_cols == n_rows-1:
        ax.set_xlabel("Time (ms from LED)")
    if idx % n_cols == 0:
        ax.set_ylabel("dF/F")

# remove unused sub-axes
for i in range(n_plots, len(axes)):
    fig.delaxes(axes[i])

fig.suptitle("Probe trials and their immediate followers", fontsize=16)
fig.tight_layout(rect=[0,0,1,0.96])
plt.savefig(pdf_out, dpi=300)
print(f"PDF saved  →  {pdf_out}")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

save_path = "/Users/omid/Downloads/E5LG_CRBL_crux2_20250430_EBC-219/saved_trials.h5"
pre_time = 100
post_time = 600
aligned_window = (-pre_time, post_time)

long_trials_dff = []
short_trials_dff = []
airpuff_windows_long = []
airpuff_windows_short = []
led_durations_long = []
led_durations_short = []
time_vectors_long = []
time_vectors_short = []

with h5py.File(save_path, "r") as f:
    trial_ids = sorted(f["trial_id"].keys(), key=lambda x: int(x))

    for trial_id in trial_ids:
        grp = f[f"trial_id/{trial_id}"]
        time = grp["time"][:]
        dff = grp["dff"][:]
        led_on = np.array(grp["LED_on"][:]).squeeze()
        led_off = np.array(grp["LED_off"][:]).squeeze()
        airpuff = grp["AirPuff"][:]

        aligned_time = time - led_on
        if len(aligned_time) != dff.shape[1]:
            print(f"Warning: Trial {int(trial_id) + 1} mismatch: trimming time/dff")
            min_len = min(len(aligned_time), dff.shape[1])
            aligned_time = aligned_time[:min_len]
            dff = dff[:, :min_len]

        mask = (aligned_time >= aligned_window[0]) & (aligned_time <= aligned_window[1])
        if np.sum(mask) == 0:
            continue

        aligned_time_cut = aligned_time[mask]
        dff_cut = dff[:, mask]

        puff_start_aligned = airpuff[0] - led_on
        puff_end_aligned = airpuff[1] - led_on
        delay = round(puff_start_aligned)

        if 390 <= delay <= 410:
            long_trials_dff.append(dff_cut)
            airpuff_windows_long.append((puff_start_aligned, puff_end_aligned))
            led_durations_long.append(led_off - led_on)
            time_vectors_long.append(aligned_time_cut)
        elif 190 <= delay <= 210:
            short_trials_dff.append(dff_cut)
            airpuff_windows_short.append((puff_start_aligned, puff_end_aligned))
            led_durations_short.append(led_off - led_on)
            time_vectors_short.append(aligned_time_cut)

def crop_and_stack(dff_trials, time_vectors):
    if not dff_trials:
        return None, None
    min_len = min(d.shape[1] for d in dff_trials)
    cropped = np.stack([d[:, :min_len] for d in dff_trials], axis=0)  # shape = (trials, neurons, time)
    time_vec = time_vectors[0][:min_len]
    return cropped, time_vec

long_data, time_long = crop_and_stack(long_trials_dff, time_vectors_long)
short_data, time_short = crop_and_stack(short_trials_dff, time_vectors_short)

# Plot function: average neurons first, then trials
def plot_avg_by_trial_first(data, time_vector, title, puff_color, puff_window, led_duration):
    if data is None:
        return
    # Step 1: average neurons (axis=1) → shape: (trials, time)
    mean_per_trial = data.mean(axis=1)

    # Step 2: average trials (axis=0) → shape: (time,)
    trace_mean = mean_per_trial.mean(axis=0)
    trace_sem = mean_per_trial.std(axis=0) / np.sqrt(mean_per_trial.shape[0])

    # Plot
    plt.figure(figsize=(8, 4))
    plt.plot(time_vector, trace_mean, color='black', label='Mean dF/F (trials)')
    plt.fill_between(time_vector, trace_mean - trace_sem, trace_mean + trace_sem, color='gray', alpha=0.3)

    plt.axvspan(0, led_duration, color='gray', alpha=0.2, label='LED')
    puff_start, puff_end = puff_window
    plt.axvspan(puff_start, puff_end, color=puff_color, alpha=0.3, label='AirPuff')

    plt.title(title)
    plt.xlabel("Time from LED onset (ms)")
    plt.ylabel("dF/F")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Existing function: average trials first, then neurons
def plot_avg_by_neuron_first(data, time_vector, title, puff_color, puff_window, led_duration):
    if data is None:
        return
    mean_per_neuron = data.mean(axis=0)
    trace_mean = mean_per_neuron.mean(axis=0)
    trace_sem = mean_per_neuron.std(axis=0) / np.sqrt(mean_per_neuron.shape[0])

    plt.figure(figsize=(8, 4))
    plt.plot(time_vector, trace_mean, color='black', label='Mean dF/F (neurons)')
    plt.fill_between(time_vector, trace_mean - trace_sem, trace_mean + trace_sem, color='gray', alpha=0.3)

    plt.axvspan(0, led_duration, color='gray', alpha=0.2, label='LED')
    puff_start, puff_end = puff_window
    plt.axvspan(puff_start, puff_end, color=puff_color, alpha=0.3, label='AirPuff')

    plt.title(title)
    plt.xlabel("Time from LED onset (ms)")
    plt.ylabel("dF/F")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot both styles for long-delay
if long_data is not None:
    median_led_long = np.median(led_durations_long)
    median_puff_long = np.median(airpuff_windows_long, axis=0)
    plot_avg_by_trial_first(long_data, time_long, "Long-delay: Avg Neurons → Avg Trials", "green", median_puff_long, median_led_long)
    plot_avg_by_neuron_first(long_data, time_long, "Long-delay: Avg Trials → Avg Neurons", "green", median_puff_long, median_led_long)

# Plot both styles for short-delay
if short_data is not None:
    median_led_short = np.median(led_durations_short)
    median_puff_short = np.median(airpuff_windows_short, axis=0)
    plot_avg_by_trial_first(short_data, time_short, "Short-delay: Avg Neurons → Avg Trials", "blue", median_puff_short, median_led_short)
    plot_avg_by_neuron_first(short_data, time_short, "Short-delay: Avg Trials → Avg Neurons", "blue", median_puff_short, median_led_short)
```

```python
import os, h5py, numpy as np, matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.utils import shuffle

session_dir = "/Users/omid/Downloads/E5LG_CRBL_crux1_20250516_EBC-325"
h5_file     = os.path.join(session_dir, "saved_trials.h5")

pdf_epoch   = os.path.join(session_dir, "epoch_averages.pdf")
pdf_decode  = os.path.join(session_dir, "svm_decoding.pdf")

pre_time, post_time = 100, 600           # ms window around LED
aligned_window      = (-pre_time, post_time)

blocks = [                                 # (start-id, end-id, label, puff-color)
    (0,   64,  "Long₁",  "lime"),
    (65, 119,  "Short",  "blue"),
    (120,149, "Long₂",   "lime"),
]

EARLY_TRIALS = 10          # if block shorter than 2× this → halves

def classify_delay(delay_ms):
    """short ~200 ms, long ~400 ms; returns ('short','long', or None)."""
    if 190 <= delay_ms <= 210:
        return "short"
    if 390 <= delay_ms <= 410:
        return "long"
    return None

def align_trim(time_raw, dff_raw, led_on):
    """Align to LED and trim time/dff mismatch, then window to –100…600 ms."""
    t = time_raw - led_on
    if t.size != dff_raw.shape[1]:
        m = min(t.size, dff_raw.shape[1])
        t, dff_raw = t[:m], dff_raw[:, :m]
    mask = (t >= aligned_window[0]) & (t <= aligned_window[1])
    if mask.sum() == 0:
        return None, None
    return t[mask], dff_raw[:, mask]

block_trials = {lbl: [] for _,_,lbl,_ in blocks}     # each item: (trial_id, t_vec, dff)
with h5py.File(h5_file, "r") as f:
    for tid_str in sorted(f["trial_id"].keys(), key=int):
        tid = int(tid_str)
        grp = f[f"trial_id/{tid}"]
        t_raw = grp["time"][:]
        dff   = grp["dff"][:]
        led   = grp["LED_on"][...].squeeze()
        puff  = grp["AirPuff"][0]

        # align + window
        t_al, d_al = align_trim(t_raw, dff, led)
        if t_al is None:
            continue

        # stash in the right block
        for start,end,lbl,_ in blocks:
            if start <= tid <= end:
                block_trials[lbl].append((tid, t_al, d_al))
                break

def mean_sem(trial_list):
    """returns mean (nTime,), sem (nTime,), common time vector (nTime,)"""
    min_len = min(d.shape[1] for _,_,d in trial_list)
    data = np.stack([d[:, :min_len] for _,_,d in trial_list])    # (T, R, time)
    mean = data.mean(axis=0).mean(axis=0)                       # avg over R & T
    sem  = data.mean(axis=0).std(axis=0) / np.sqrt(data.shape[0])
    t    = trial_list[0][1][:min_len]
    return mean, sem, t

with PdfPages(pdf_epoch) as pdf:
    for start,end,lbl,puff_col in blocks:
        trials = sorted(block_trials[lbl], key=lambda x: x[0])
        n = len(trials)
        if n == 0:
            continue

        # Early / Late selection
        if n >= 2*EARLY_TRIALS:
            early = trials[:EARLY_TRIALS]
            late  = trials[-EARLY_TRIALS:]
        else:                       # fall-back to halves
            early = trials[: n//2]
            late  = trials[n//2 :]

        m_e, s_e, t_e = mean_sem(early)
        m_l, s_l, t_l = mean_sem(late)
        L = min(t_e.size, t_l.size)
        m_e, s_e, t_e = m_e[:L], s_e[:L], t_e[:L]
        m_l, s_l, t_l = m_l[:L], s_l[:L], t_l[:L]   # now t_e == t_l

        # plot
        fig, ax = plt.subplots(figsize=(6,4))
        ax.plot(t_e, m_e, color="black", label=f"Early ({len(early)} trials)")
        ax.fill_between(t_e, m_e-s_e, m_e+s_e, color="black", alpha=.25)
        ax.plot(t_e, m_l, color="red",   label=f"Late ({len(late)} trials)")
        ax.fill_between(t_e, m_l-s_l, m_l+s_l, color="red", alpha=.25)

        ax.axvspan(0, 50, color="gray", alpha=.3, label="LED")
        puff_start = 200 if puff_col=="blue" else 400
        ax.axvspan(puff_start, puff_start+20, color=puff_col, alpha=.3, label="Air-puff")

        ax.set_xlim(aligned_window)
        ax.set_xlabel("Time from LED onset (ms)")
        ax.set_ylabel("dF/F (mean ± SEM)")
        ax.set_title(f"{lbl}: Early vs Late")
        ax.legend()
        fig.tight_layout()
        pdf.savefig(fig, dpi=300)
        plt.close(fig)

print(f"✅ epoch_averages.pdf saved → {pdf_epoch}")
```

```python
# HELPER: SVM time-course per block
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.svm import SVC

def svm_timecourse_holdout(trial_list, labels, n_splits=200, test_size=0.20):
    """
    trial_list : list of (tid, tvec, dff) already trimmed to common length
    labels     : length = nTrials   (0 = Early, 1 = Late)

    Returns  mean_acc (nTime,), sem_acc (nTime,), time_vector
    """
    X = np.stack([d for _,_,d in trial_list])          # (T, R, time)
    y = np.array(labels)
    nT, nR, nTime = X.shape

    acc_all = np.zeros((n_splits, nTime))
    sss = StratifiedShuffleSplit(
        n_splits=n_splits, test_size=test_size, random_state=0
    )

    for split_idx, (train, test) in enumerate(sss.split(X, y)):
        for t in range(nTime):
            clf = SVC(kernel="linear")
            clf.fit(X[train, :, t], y[train])
            acc_all[split_idx, t] = clf.score(X[test, :, t], y[test])

    mean_acc = acc_all.mean(axis=0)
    sem_acc  = acc_all.std(axis=0) / np.sqrt(n_splits)
    time_vec = trial_list[0][1][:nTime]
    return mean_acc, sem_acc, time_vec
```

```python
import pathlib, sys, h5py, numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster        import KMeans
from sklearn.metrics        import silhouette_score
from sklearn.preprocessing  import StandardScaler

# ----------------------------------------------------------------------
#  paths & constants
# ----------------------------------------------------------------------
ROOT      = pathlib.Path("/Volumes/T9/neuraldata/NajafiLab")
SESS_DIR  = ROOT / "E5LG_CRBL_crux1_20250516_EBC-325"
H5_FILE   = SESS_DIR / "saved_trials.h5"

PRE_MS    = 100        # window in ms (before LED_on)
POST_MS   = 600        # window in ms (after  LED_on)
WIN       = (-PRE_MS, POST_MS)        # inclusive

PDF_SCAT  = SESS_DIR / "pca_scatter.pdf"
PDF_TRACES= SESS_DIR / "cluster_traces.pdf"

# ----------------------------------------------------------------------
#  helper – load ONE trial, robust to length mismatch
# ----------------------------------------------------------------------
def load_one_trial(grp):
    """
    Return (t_aligned_window , dff_window).

    • Trims *time* and *dff* to the SAME minimum length when they differ.
    • Keeps only samples inside WIN = (-PRE_MS, POST_MS).
    • If that window is empty, returns (None, None) so the caller can skip.
    """
    time      = grp["time"][:]                 # (T_raw,)
    dff       = grp["dff"][:]                  # (neurons × T_raw)
    led_on    = np.array(grp["LED_on"][:]).squeeze()

    t_aligned = time - led_on                  # (T_raw,)

    # ----------- harmonise lengths ------------------------------------
    min_len   = min(len(t_aligned), dff.shape[1])
    if len(t_aligned) != dff.shape[1]:
        t_aligned = t_aligned[:min_len]
        dff       = dff[:, :min_len]

    # ----------- apply analysis window --------------------------------
    keep      = (t_aligned >= WIN[0]) & (t_aligned <= WIN[1])
    if not keep.any():
        return None, None                      # nothing in window → skip

    return t_aligned[keep], dff[:, keep]


# ----------------------------------------------------------------------
#  helper – average across all trials that have data in the window
# ----------------------------------------------------------------------
def trial_averages(h5_path: pathlib.Path):
    """
    Return (time_ms , mean_dff) where

        time_ms   … 1‑D vector of length T (shared across trials)
        mean_dff  … (neurons × T) matrix, mean of every usable trial

    Trials whose aligned data fall completely outside WIN are skipped.
    """
    traces   = []                       # list of (neurons × T_i) arrays
    time_ms  = None                     # filled by first valid trial

    with h5py.File(h5_path, "r") as f:
        trial_ids = sorted(f["trial_id"].keys(), key=int)

        for tid in trial_ids:
            t, d = load_one_trial(f[f"trial_id/{tid}"])
            if t is None:               # window empty  →  skip trial
                continue
            traces.append(d)
            if time_ms is None:         # first good trial sets the clock
                time_ms = t

    if not traces:
        raise RuntimeError("No trials contained samples inside WIN")

    # -------- make every trial the same length -------------------------
    min_T   = min(d.shape[1] for d in traces)
    traces  = [d[:, :min_T] for d in traces]
    time_ms = time_ms[:min_T]

    # -------- average --------------------------------------------------
    mean_dff = np.stack(traces, axis=0).mean(axis=0)   # trials → mean
    return time_ms, mean_dff


def pick_k(data, k_max: int = 12):
    """Return best K via highest Silhouette; also print a mini‑report."""
    best_k, best_score = 2, -1
    for k in range(2, min(k_max, len(data))):
        labs   = KMeans(n_clusters=k, n_init="auto", random_state=1).fit_predict(data)
        score  = silhouette_score(data, labs)
        print(f"K={k:2d}  silhouette={score:.4f}")
        if score > best_score:
            best_k, best_score = k, score
    print(f"→ chosen K = {best_k}\n")
    return best_k

# ----------------------------------------------------------------------
#  main
# ----------------------------------------------------------------------
def main():
    if not H5_FILE.exists():
        sys.exit(f"H5 file not found:\n{H5_FILE}")

    print("Loading & averaging trials …")
    time_ms, dff = trial_averages(H5_FILE)          # neurons × T
    n_neurons, T = dff.shape
    print(f"{n_neurons} neurons · {T} time points\n")

    # ------------------------------------------------------------------
    #  build feature matrix  (neurons × T) → (neurons × T)
    # ------------------------------------------------------------------
    X = dff.copy()                  # each row = one neuron’s trace

    # z‑score across time for every neuron
    scaler = StandardScaler(with_mean=True, with_std=True)
    X_z    = scaler.fit_transform(X)

    # ------------------------------------------------------------------
    #  PCA  (95 % variance)
    # ------------------------------------------------------------------
    pca     = PCA(n_components=0.95, svd_solver="full", random_state=0)
    X_pca   = pca.fit_transform(X_z)    # neurons × n_PC
    print(f"PCA kept {X_pca.shape[1]} PCs (95 % variance)\n")

    # ------------------------------------------------------------------
    #  pick K & cluster
    # ------------------------------------------------------------------
    K       = pick_k(X_pca, k_max=12)
    kmodel  = KMeans(n_clusters=K, n_init="auto", random_state=1)
    labels  = kmodel.fit_predict(X_pca)

    # ------------------------------------------------------------------
    #  plot 1: PCA scatter
    # ------------------------------------------------------------------
    plt.figure(figsize=(6, 5))
    for k in range(K):
        sel = labels == k
        plt.scatter(X_pca[sel, 0], X_pca[sel, 1], s=10, label=f"C{k}")
    plt.xlabel("PC 1"); plt.ylabel("PC 2")
    plt.title("Neurons – first two PCs")
    plt.legend(markerscale=2, fontsize="small")
    plt.tight_layout()
    plt.savefig(PDF_SCAT, dpi=300)
    plt.close()
    print(f"saved ⇒ {PDF_SCAT}")

    # ------------------------------------------------------------------
    #  plot 2: mean ± SEM trace per cluster
    # ------------------------------------------------------------------
    cols = 3
    rows = int(np.ceil(K / cols))
    fig, axes = plt.subplots(rows, cols,
                             figsize=(cols * 4, rows * 3),
                             sharex=True, sharey=True)
    axes = axes.flatten()

    for k in range(K):
        ax        = axes[k]
        neurons_k = dff[labels == k]      # n_k × T
        if neurons_k.size == 0:
            ax.set_axis_off(); continue
        mean_k    = neurons_k.mean(axis=0)
        sem_k     = neurons_k.std(axis=0) / np.sqrt(neurons_k.shape[0])
        ax.plot(time_ms, mean_k, lw=1.2)
        ax.fill_between(time_ms, mean_k - sem_k, mean_k + sem_k,
                        color="gray", alpha=0.3)
        ax.axvspan(0, 50, color="lightgray", alpha=0.3)     # LED window
        ax.set_title(f"Cluster {k}  (n={neurons_k.shape[0]})")
        if k % cols == 0:
            ax.set_ylabel("dF/F")
    for ax in axes[K:]:
        ax.set_axis_off()

    fig.suptitle("Cluster mean traces", fontsize=14)
    fig.tight_layout(rect=[0, 0, 1, 0.93])
    fig.savefig(PDF_TRACES, dpi=300)
    plt.close(fig)
    print(f"saved ⇒ {PDF_TRACES}\nDone.")

# ----------------------------------------------------------------------
if __name__ == "__main__":
    main()
```

```python
import pathlib, sys, h5py, numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing  import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster        import KMeans
from sklearn.metrics        import silhouette_score


# ----------------------------------------------------------------------
#  CONSTANTS ––– tune here only! –––––––––––––––––––––––––––––––––––––––
# ----------------------------------------------------------------------
ROOT       = pathlib.Path("/Volumes/T9/neuraldata/NajafiLab")
SESS_DIR   = ROOT / "E5LG_CRBL_crux1_20250516_EBC-325"
H5_FILE    = SESS_DIR / "saved_trials.h5"

FS_HZ      = 30                     # imaging frame rate
PRE_MS     = 200                    # window before LED
POST_MS    = 800                    # window after  LED
WIN        = (-PRE_MS, POST_MS)     # inclusive

LAG_MS     = 800                    # kernel length
N_LAGS     = int(round(LAG_MS / (1000/FS_HZ)))     # ≈ 24–26 frames
RIDGE_LAMB = 5.0                    # λ for ridge

PCA_VAR    = 0.95                   # 95 % var. retained
K_MAX      = 15                     # search upper bound for K

PDF_SCAT   = SESS_DIR / "glm_pca_scatter.pdf"
PDF_TRACES = SESS_DIR / "glm_cluster_traces.pdf"


# ----------------------------------------------------------------------
#  SMALL HELPERS
# ----------------------------------------------------------------------
def make_toeplitz(event_vec: np.ndarray, n_lags: int):
    """
    Build a lagged design matrix for *one* binary event column.

    event_vec … shape (T,) with 0/1 entries
    returns   … shape (T , n_lags) ; lag 0 = original column
    """
    T = len(event_vec)
    X = np.zeros((T, n_lags), dtype=float)
    for i in range(n_lags):
        if i == 0:
            X[:, 0] = event_vec
        else:
            X[i:, i] = event_vec[:-i]
    return X


def design_matrix(t_ms: np.ndarray):
    """
    3 event types  ×  N_LAGS  →  shape (T , 3·N_LAGS)
    """
    led         = ((t_ms >=   0) & (t_ms <=  50)).astype(float)

    puff_short  = ((t_ms >= 190) & (t_ms <= 210)).astype(float)
    puff_long   = ((t_ms >= 390) & (t_ms <= 410)).astype(float)

    X = np.hstack([
        make_toeplitz(led,        N_LAGS),
        make_toeplitz(puff_short, N_LAGS),
        make_toeplitz(puff_long,  N_LAGS),
    ])
    return X


def ridge_glm(X: np.ndarray, y: np.ndarray, lam: float):
    """
    Closed‑form ridge:  β = (XᵀX + λI)⁻¹ Xᵀy
    Returns β  (shape (P,))
    """
    P = X.shape[1]
    reg = lam * np.eye(P)
    beta = np.linalg.solve(X.T @ X + reg, X.T @ y)
    return beta


# ----------------------------------------------------------------------
#  DATA LOADING
# ----------------------------------------------------------------------
def load_one_trial(grp):
    """
    Align one trial to LED_on, trim mismatches, crop to WIN.

    Returns (t_window , dff_window) or (None , None) if window empty.
    """
    time   = grp["time"][:]                 # (T_raw,)
    dff    = grp["dff"][:]                  # (neurons × T_raw)
    led_on = np.array(grp["LED_on"][:]).squeeze()

    t_al   = time - led_on                  # align

    # harmonise lengths -------------------------------------------------
    min_len = min(len(t_al), dff.shape[1])
    if len(t_al) != dff.shape[1]:
        t_al = t_al[:min_len]
        dff  = dff[:, :min_len]

    # crop to analysis window ------------------------------------------
    keep = (t_al >= WIN[0]) & (t_al <= WIN[1])
    if not keep.any():                      # nothing for this trial
        return None, None

    return t_al[keep], dff[:, keep]


def trial_averages(h5_path: pathlib.Path):
    """
    Returns: time_ms (T,) , mean_dff (neurons × T)

    Only trials with samples inside WIN are used.
    """
    traces  = []               # list of (neurons × T_i)
    time_ms = None

    with h5py.File(h5_path, "r") as f:
        tids = sorted(f["trial_id"].keys(), key=int)
        for tid in tids:
            t, d = load_one_trial(f[f"trial_id/{tid}"])
            if t is None:
                continue
            traces.append(d)
            if time_ms is None:
                time_ms = t

    if not traces:
        raise RuntimeError("No trial contains data inside WIN")

    min_T   = min(d.shape[1] for d in traces)
    traces  = [d[:, :min_T] for d in traces]
    time_ms = time_ms[:min_T]

    mean_dff = np.stack(traces, axis=0).mean(axis=0)   # mean over trials
    return time_ms, mean_dff


# ----------------------------------------------------------------------
#  AUTO‑SELECT K
# ----------------------------------------------------------------------
def pick_k(data, k_max=K_MAX):
    best_k, best_score = 2, -1
    for k in range(2, min(k_max, len(data))):
        labs  = KMeans(n_clusters=k, n_init="auto", random_state=1).fit_predict(data)
        score = silhouette_score(data, labs)
        print(f"K={k:2d}  silhouette={score:.4f}")
        if score > best_score:
            best_k, best_score = k, score
    print(f"→ chosen K = {best_k}\n")
    return best_k


# ----------------------------------------------------------------------
#  MAIN
# ----------------------------------------------------------------------
def main():
    if not H5_FILE.exists():
        sys.exit(f"H5 file not found:\n{H5_FILE}")

    print("Loading & averaging trials …")
    time_ms, dff = trial_averages(H5_FILE)          # neurons × T
    n_neurons, T = dff.shape
    print(f"{n_neurons} neurons · {T} frames (aligned)\n")

    # ------------------------------------------------------------------
    #  build GLM feature matrix  (neurons × P)
    # ------------------------------------------------------------------
    X_design = design_matrix(time_ms)               # (T , 3·N_LAGS)
    P        = X_design.shape[1]
    features = np.empty((n_neurons, P), dtype=float)

    for i in range(n_neurons):
        y_i        = dff[i]                         # (T,)
        beta_i     = ridge_glm(X_design, y_i, RIDGE_LAMB)
        features[i] = beta_i                       # row = concatenated kernels

    # z‑score across features -------------------------------------------------
    scaler  = StandardScaler()
    Z_feat  = scaler.fit_transform(features)

    # ------------------------------------------------------------------
    #  PCA
    # ------------------------------------------------------------------
    pca   = PCA(n_components=PCA_VAR, svd_solver="full", random_state=0)
    PCs   = pca.fit_transform(Z_feat)              # neurons × n_PC
    print(f"PCA kept {PCs.shape[1]} components (95 % var)\n")

    # ------------------------------------------------------------------
    #  select K  &  cluster
    # ------------------------------------------------------------------
    K      = pick_k(PCs, k_max=K_MAX)
    kmodel = KMeans(n_clusters=K, n_init="auto", random_state=1)
    labels = kmodel.fit_predict(PCs)

    # ------------------------------------------------------------------
    #  PLOT 1 – PCA scatter
    # ------------------------------------------------------------------
    plt.figure(figsize=(6, 5))
    for k in range(K):
        sel = labels == k
        plt.scatter(PCs[sel, 0], PCs[sel, 1], s=12, label=f"C{k}")
    plt.xlabel("PC 1"); plt.ylabel("PC 2")
    plt.title("GLM‑kernel PCA (first two PCs)")
    plt.legend(markerscale=2, fontsize="small")
    plt.tight_layout()
    plt.savefig(PDF_SCAT, dpi=300)
    plt.close()
    print(f"saved ⇒ {PDF_SCAT}")

    # ------------------------------------------------------------------
    #  PLOT 2 – mean ± SEM trace per cluster
    # ------------------------------------------------------------------
    cols = 3
    rows = int(np.ceil(K / cols))
    fig, axes = plt.subplots(rows, cols,
                             figsize=(cols * 4, rows * 3),
                             sharex=True, sharey=True)
    axes = axes.flatten()

    for k in range(K):
        ax        = axes[k]
        sel       = labels == k
        neurons_k = dff[sel]                       # n_k × T
        if neurons_k.size == 0:
            ax.set_axis_off(); continue
        mean_k = neurons_k.mean(axis=0)
        sem_k  = neurons_k.std(axis=0) / np.sqrt(neurons_k.shape[0])
        ax.plot(time_ms, mean_k, lw=1.2)
        ax.fill_between(time_ms, mean_k - sem_k, mean_k + sem_k,
                        color="gray", alpha=0.3)
        ax.axvspan(0, 50, color="lightgray", alpha=0.3)
        ax.set_title(f"Cluster {k}  (n={neurons_k.shape[0]})")
        if k % cols == 0:
            ax.set_ylabel("dF/F")
    for ax in axes[K:]:
        ax.set_axis_off()

    fig.suptitle("Clusters – mean dF/F traces", fontsize=14)
    fig.tight_layout(rect=[0, 0, 1, 0.93])
    fig.savefig(PDF_TRACES, dpi=300)
    plt.close(fig)
    print(f"saved ⇒ {PDF_TRACES}\nDone.")


# ----------------------------------------------------------------------
if __name__ == "__main__":
    main()
```
